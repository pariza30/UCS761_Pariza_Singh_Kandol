```python
import numpy as np
```

```python
# Step activation function
def step(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
# Perceptron function
def perceptron(x1, x2, w1, w2, b):
    z = w1*x1 + w2*x2 + b
    return step(z)
```

```python
inputs = [(0,0),(0,1),(1,0),(1,1)]
```

```python
print("AND Gate")
for x1,x2 in inputs:
    print(x1,x2,"->",perceptron(x1,x2,1,1,-1.5))
```

```python
print("\nOR Gate")
for x1,x2 in inputs:
    print(x1,x2,"->",perceptron(x1,x2,1,1,-0.5))
```

```python
print("\nNAND Gate")
for x1,x2 in inputs:
    print(x1,x2,"->",perceptron(x1,x2,-1,-1,1.5))
```

```python
print("\nNOR Gate")
for x1,x2 in inputs:
    print(x1,x2,"->",perceptron(x1,x2,-1,-1,0.5))
```
